### Clean up the data!

#### Drop rows of NAN

```python
## checking to make sure the nan is actually empty rows
# all rows with nan
nan_df = all_data[all_data.isna().any(axis=1)]
print(nan_df)
# comparing to rows with nan under qty ordered only
test2 = all_data[all_data['Quantity Ordered'].isna()]
print(test2)

print('----------')

# removing the nan rows
all_data = all_data.dropna(how='all')
# new_df[new_df['Quantity Ordered'].isna()]
all_data.count
```

#### Find 'Or' and delete it

```python
import re
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
# print(temp_df)
all_data.count
all_data.head()
# temp_df.count
# all_data.drop("^Or")
```

#### Convert columns to the correct type

```python
## keith recommended .to_numeric...says it's more stable or just better for pandas in general
# all_data['Quantity Ordered'] = all_data['Quantity Ordered'].astype(int)
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
# all_data['Price Each'] = all_data['Price Each'].astype(float)
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

```python
# len(all_data['Price Each'])
# all_data.insert(5, "Order Month", ['Test']*len(all_data['Price Each']))
all_data['Order Month'] = np.NaN
all_data.head()

```

#### Task 3: Add a sales column

```python
all_data['Total Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

#### Question 1: What was the best month for sales? How much was earned that month?

```python
# own attempt

sales_data = all_data.groupby('Order Month', as_index=False).sum()
sales_data.head()
```

```python
import matplotlib.pyplot as plt

plt.title('Total Sales per Month')
plt.xlabel('Month')
plt.ylabel('Sales in USD ($)')
plt.xticks(np.arange(0, 13, step=1))
plt.bar(sales_data['Order Month'], sales_data['Total Sales'])
plt.show()
```

#### Question 2: What city had the highest number of sales?

#### Task 4: Add a city column

```python
# my solution
# all_data['City'] = all_data['Purchase Address'].str.split(', ').str[1]
# all_data.head()



# keith's way of doing it using .apply
def get_city(address):
    return f"{address.split(', ')[1]}, {address.split(', ')[2].split(' ')[0]}"

# all_data['City'] = all_data['Purchase Address'].apply(lambda x: x.split(', ')[1] + ', ' + x.split(', ')[2].split(' ')[0])
all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x))
all_data.head()
```

```python
city_data = all_data.groupby('City', as_index=False).sum()
city_data.head()
```

```python
plt.title('Total Sales by City')
plt.xlabel('City')
plt.xticks(rotation=90, size=8)
plt.ylabel('Sales in USD ($)')
plt.bar(city_data['City'], city_data['Total Sales'])
plt.show()
```

#### Question 4: What products are most often sold together?

```python
# list of orders w/ multiple orders

# doing it my way initially
all_data.head()
by_order_id = all_data.groupby('Order ID', as_index=False).count()
by_order_id = by_order_id[by_order_id['Product'] > 1]
by_order_id.head()

# mult_orders df with all_data filtered down to only mult items in order
all_data.head()
mult_orders = all_data[all_data['Order ID'].isin(by_order_id['Order ID'])]
mult_orders.head()

```

```python
# using pd.duplicated()
mult_orders = all_data[all_data['Order ID'].duplicated(keep=False)]
mult_orders.head()
```

```python
# grouping orders together and put products on the same line in new column

mult_orders['Grouped'] = mult_orders.groupby('Order ID')['Product'].transform(lambda x: ', '.join(x))
mult_orders.head()
```

```python
# drop duplicated Grouped for same order id

# mult_orders = mult_orders[mult_orders['Order ID'].duplicated()]
# mult_orders.head()

# OR

mult_orders = mult_orders[['Order ID', 'Grouped']].drop_duplicates()
mult_orders.head()
```

```python
# count the pairs of what occurs together
# https://stackoverflow.com/questions/52195887/counting-unique-pairs-of-numbers-into-a-python-dictionary
from itertools import combinations
from collections import Counter

count = Counter()

for row in mult_orders['Grouped']:
    row_list = row.split(', ')
    count.update(Counter(combinations(row_list, 2)))
    
for key, value in count.most_common(10):
    print(key, value)
```

#### Question 5: What product sold the most? Why do you think it sold the most?

```python
all_data.head()
```

```python
product_group = all_data[['Product', 'Quantity Ordered']].groupby('Product', as_index=False).sum()#.sort_values('Quantity Ordered', ascending=False)
product_group.head()
```

```python
plt.title('Most Popular Products')
plt.xlabel('Product')
plt.ylabel('Quantity Ordered')
plt.xticks(rotation=90, size=8)
plt.grid()
plt.bar(product_group['Product'], product_group['Quantity Ordered'])
plt.show()
```

```python
prices = all_data[['Product', 'Price Each']].groupby('Product', as_index=False).mean()
prices.head()
```
